```python
import pandas as pd 
import plotly.express as px
c_names =["pregnant","glucose","bp","skin", "insulin","bmi","pedigree","age","label"]
df = pd.read_csv("data_119.csv", names = c_names).iloc[1:]

print(df.head())
```

```python
features = ["pregnant","insulin","bmi","age","glucose","bp","pedigree"]
X = df[features]
y = df.label
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 1)

clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy", metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus

dot_data = StringIO()
export_graphviz(clf,out_file = dot_data, filled = True, special_characters = True, feature_names = features, class_names=["0", "1"])

print(dot_data.getvalue())
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png("diabetes.png")
Image(graph.create_png())
```
